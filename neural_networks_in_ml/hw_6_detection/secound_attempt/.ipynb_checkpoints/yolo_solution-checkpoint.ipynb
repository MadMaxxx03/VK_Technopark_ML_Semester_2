```python
import os
import cv2
import torch
import numpy as np
from torch.utils.data import Dataset, DataLoader
from torch.utils.data.dataloader import default_collate
from torchvision import transforms
import torch.optim as optim
from tqdm import tqdm
import matplotlib.pyplot as plt
import matplotlib.patches as patches
%matplotlib inline
import random
```

```python
model = torch.hub.load('ultralytics/yolov5', 'yolov5s', pretrained=True)

# Изменяем количество выходных классов с 80 на 1 (одноклассовая детекция)
model.model.nc = 1  # Указываем количество классов (1 класс)

# Пересчитываем веса для одноклассовой детекции
model.model.class_weights = None  # Сброс весов классов

# Пересчитаем веса для одного класса (без имени класса, если не нужно)
if hasattr(model.model, 'initialize_biases'):
    model.model.initialize_biases()  # Инициализация весов

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
def plot_image_with_boxes(image_path, label_path):

    image = cv2.imread(image_path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Конвертация BGR -> RGB

    h, w, _ = image.shape

    with open(label_path, 'r') as f:
        labels = f.readlines()

    for label in labels:
        _, x_center, y_center, width, height = map(float, label.strip().split())  # class_id игнорируем
        
        x_center *= w
        y_center *= h
        width *= w
        height *= h
        
        x1 = int(x_center - width / 2)
        y1 = int(y_center - height / 2)
        x2 = int(x_center + width / 2)
        y2 = int(y_center + height / 2)

        image = cv2.rectangle(image, (x1, y1), (x2, y2), color=(255, 0, 0), thickness=2)

    plt.figure(figsize=(10, 10))
    plt.imshow(image)
    plt.axis('off')
    plt.show()
```

```python
class RoadSignsDataset(Dataset):
    def __init__(self, images_dir, labels_dir, transform=None):

        self.images_dir = images_dir
        self.labels_dir = labels_dir
        self.transform = transform
        self.image_filenames = os.listdir(images_dir)

    def __len__(self):
        return len(self.image_filenames)

    def __getitem__(self, idx):
        img_filename = self.image_filenames[idx]
        img_path = os.path.join(self.images_dir, img_filename)
        
        label_filename = img_filename.replace('.jpg', '.txt')  # предполагаем, что разметки имеют одинаковые имена
        label_path = os.path.join(self.labels_dir, label_filename)

        image = cv2.imread(img_path)
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

        boxes = []
        with open(label_path, 'r') as f:
            labels = f.readlines()
            for label in labels:
                _, x_center, y_center, width, height = map(float, label.strip().split())
                boxes.append([x_center, y_center, width, height])

        boxes = torch.tensor(boxes, dtype=torch.float32)

        if self.transform:
            transformed = self.transform(image=image, bboxes=boxes)
            image = transformed['image']
            boxes = transformed['bboxes']

        image = torch.tensor(image, dtype=torch.float32).permute(2, 0, 1) / 255.0  # Преобразуем из HWC в CHW и нормализуем

        return image, boxes
```

```python
images_dir = 'data/train/images'
labels_dir = 'data/train/labels'
val_images_dir = 'data/valid/images'
val_labels_dir = 'data/valid/labels'

dataset = RoadSignsDataset(images_dir, labels_dir)

num_samples = len(dataset)
num_samples_10_percent = int(num_samples * 0.1)

subset_indices = list(range(num_samples_10_percent))  # Индексы для первых 10%
subset_dataset = torch.utils.data.Subset(dataset, subset_indices)

dataloader = DataLoader(subset_dataset, batch_size=8, shuffle=True)

for images, boxes in dataloader:
    print(images.shape)  
    print(boxes)         
    break  
```

```python
def visualize_predictions(model, images):

    image = images[1].unsqueeze(0)  
    
    model.eval()

    with torch.no_grad():
        results = model(image) 

    predictions = results[0]  

    bboxes = predictions[:, :4]  
    scores = predictions[:, 4]   
    classes = predictions[:, 5]  

    confidence_threshold = 0.1
    conf_mask = scores >= confidence_threshold
    bboxes = bboxes[conf_mask]
    scores = scores[conf_mask]
    classes = classes[conf_mask]

    nms_indices = torch.ops.torchvision.nms(bboxes, scores, iou_threshold=0.4)
    bboxes = bboxes[nms_indices]
    scores = scores[nms_indices]
    classes = classes[nms_indices]

    image_np = image.squeeze(0).permute(1, 2, 0).cpu().numpy()  

    fig, ax = plt.subplots(1)
    ax.imshow(image_np)

    for bbox, score, class_id in zip(bboxes, scores, classes):
        x_min, y_min, x_max, y_max = bbox
        x_min, y_min, x_max, y_max = x_min.item(), y_min.item(), x_max.item(), y_max.item()

        rect = patches.Rectangle((x_min, y_min), x_max - x_min, y_max - y_min, linewidth=2, edgecolor='r', facecolor='none')
        ax.add_patch(rect)

        ax.text(x_min, y_min, f'{score:.2f}', color='white', fontsize=12, bbox=dict(facecolor='red', alpha=0.5))

    plt.show()
```

```python
visualize_predictions(model, images)
```

```python
def custom_yolo_loss(pred_boxes, target_boxes):
    # Преобразуем список target_boxes в тензор
    target_boxes = torch.cat(target_boxes, dim=0)
    
    # Убедимся, что размеры тензоров совпадают
    if pred_boxes.shape < target_boxes.shape:
        target_boxes = target_boxes[:8]
    if pred_boxes.shape > target_boxes.shape:
        pred_boxes = pred_boxes[:target_boxes.shape[0]]
    
    # Применяем MSE для координат боксов
    loss_location = torch.mean((pred_boxes - target_boxes) ** 2)
    
    return loss_location
```

```python
def train_loop(model, dataloader, optimizer, epochs, device):
    model.train()
    
    for epoch in range(epochs):
        running_loss = 0.0
        
        for images, target_boxes in dataloader:
            images = images.to(device)
            target_boxes = [boxes.to(device) for boxes in target_boxes]

            optimizer.zero_grad()
            
            results = model(images)
            predictions = results[0]  

            batch_size = predictions.shape[0]
            filtered_predictions = []

            for i in range(batch_size):
                preds_per_image = predictions[i]  
                box_coords = preds_per_image[..., :4] 

                filtered_predictions.append(box_coords[-1][-1][-1])

            filtered_predictions = torch.stack(filtered_predictions).to(device)  
            
            loss = custom_yolo_loss(filtered_predictions, target_boxes)

            
            # Обратное распространение и шаг оптимизатора
            #loss.backward()
            optimizer.step()
            
            running_loss += loss.item()
        
        print(f'Epoch {epoch+1}/{epochs}, Loss: {running_loss/len(dataloader)}')
```

```python
epochs=2
optimizer = optim.Adam(model.parameters(), lr=0.001)
train_loop(model, dataloader, optimizer, epochs, device)
```
